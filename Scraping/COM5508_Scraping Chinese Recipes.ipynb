```python
#创建urls.json文件
f1 = open('urls.json', 'w', encoding='utf-8')
url = 'https://www.meishij.net/china-food/caixi/'
response = request.urlopen(url)
result = response.read().decode('utf-8')
tree = etree.HTML(result)
#将大类存储到列表里
#category_list = tree.xpath('//*[@id="listnav_con_c"]/dl/dt/text()')

data = []
category_list = tree.xpath('//*[@id="listnav"]/div/dl/dd/a/text()')
#print(category_list)

for i in range(len(category_list)):
    #category_title = category_list[i]
    expression1 = '//*[@id="listnav"]/div/dl/dd['+ str(i + 1) +']/a/text()'
    #expression1 = '//*[@id="listnav_con_c"]/dl['+ str(i + 1) +']/dd['+ str(j + 1) +']/a/text()'
    expression2 = '//*[@id="listnav"]/div/dl/dd['+ str(i + 1) +']/a/@href'
    sub_category_title = tree.xpath(expression1)
    sub_category_url = tree.xpath(expression2)
    a = {'title': sub_category_title[0],'url': sub_category_url[0]}
    print(a)
    data.append(a)
json.dump(data, f1, ensure_ascii=False, indent=4)
f1.close()
```

```python
#外部传入保存的文件名及对应的URL地址
def get_single_page(f_name, f_url):
    url = f_url
    response = request.urlopen(url)
    result = response.read().decode('utf-8')
    tree = etree.HTML(result)
    #获取每个URL里的总页码
    try:
        totalpages = tree.xpath('//*[@id="listtyle1_w"]/div[2]/div/span[2]/form/text()[1]')
        totalpagesnumberlist = re.findall('[0-9]\d*', totalpages[0])
        totalpagesnumber = totalpagesnumberlist[0]
    except Exception:
        pass
    
    filename = f_name + '.json'
    f1 = open('/Users/heweiyuan/py/COM5508_Analytics/Group_project/new_trial_recipe/recipe_url_3/'+ str(filename), 'w', encoding='utf-8')
    data = []
    try:
        for i in range(1, int(totalpagesnumber) + 1):
            #构建完整URL地址
            url1 = url+'?&page='+str(i)
            print(url1)
            response1 = request.urlopen(url1)
            result1 = response1.read().decode('utf-8')
            tree1 = etree.HTML(result1)
            itemlist = tree1.xpath('//*[@id="listtyle1_list"]/div/@class')
            #print(itemlist)
            for i in range(1, len(itemlist)+1):
                str1 = '//*[@id="listtyle1_list"]/div['+ str(i)+']/a/div/div/div[1]/strong/text()'
                str2 = '//*[@id="listtyle1_list"]/div['+ str(i)+']/a/div/div/div[1]/span/text()'
                str3 = '//*[@id="listtyle1_list"]/div['+ str(i) +']/a/@href'
                item_title = tree1.xpath(str1)
                item_popular = tree1.xpath(str2)
                item_make = tree1.xpath(str3)
                a = { 'title': item_title[0],
                     'popular':item_popular,
                     'type': f_name,
                     'url': item_make[0]}
                data.append(a)
    #print(data)
    except Exception:
        pass
    json.dump(data, f1, ensure_ascii=False, indent=4)
    f1.close()
```

```python
f = open('/Users/heweiyuan/py/COM5508_Analytics/Group_project/new_trial_recipe/urls.json', 'r', encoding='utf-8')
data = json.loads(f.read(), encoding='utf-8')
 
if __name__ == "__main__":
    for i in range(len(data)):
        filename = data[i]['title']
        fileurl = data[i]['url']
        get_single_page(filename, fileurl)
```

```python
from urllib import request
import chardet
from bs4 import BeautifulSoup
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import csv
from lxml import etree
import json
import re
import os
```

```python
def detail_page(f_name,f_url,f_type,f_pop):
    filename = f_name + '.json'
    f1 = open('/Users/heweiyuan/py/COM5508_Analytics/Group_project/new_trial_recipe/recipe/'+ str(filename), 'w',encoding='utf-8')
    meishi_item = {}
    url = f_url
    response = request.urlopen(url)
    #response_1 = requests.get(url).content
    #result_save(meishi_item)
    print(f_url)
    main_ingredient=[]
    sup_ingredient=[]
    sup_vol =[]
    #url = f_url
    try:
        result = response.read().decode('utf-8')
        tree = etree.HTML(result)
        meishi_item['name'] = tree.xpath('//*[@id="tongji_title"]//text()')
        meishi_item['gy'] = tree.xpath('//*[@id="tongji_gy"]//text()')
        meishi_item['people'] = tree.xpath('//*[@id="tongji_rsh"]//text()')
        meishi_item['taste'] = tree.xpath('//*[@id="tongji_kw"]//text()')
        meishi_item['cooking_time'] = tree.xpath('//*[@id="tongji_prsj"]//text()')
        
        html = urlopen(url).read().decode("utf-8")
        soup = BeautifulSoup(html,'lxml')
        zhuliao_box = soup.find_all('div',attrs={'class':'yl zl clearfix'})
        for zl in zhuliao_box:
            for a in zl.find_all('a'):
                #print(a.text.strip(' '))
                main_ingredient.append(a.text.strip(' '))
        fuliao_box = soup.find_all('div',attrs={'class':'yl fuliao clearfix'})
        for fl in fuliao_box:
            for b in fl.find_all('h4'):
                #print(a.text.strip(' '))
                sup_ingredient.append(b.text.strip(' '))
            for b_1 in fl.find_all('span'):
                #print(b_1.text.strip(''))
                sup_vol.append(b_1.text.strip(''))
                #a= soup.find_all("div",attrs={'class':'measure'})
        print(main_ingredient)
        #meishi_item['title']= f_name
        meishi_item['main_ingredient']= main_ingredient
        meishi_item['sup_ingredient']= sup_ingredient
        meishi_item['sup_vol']= sup_vol
        meishi_item['type']= [f_type]
        meishi_item['popular']= f_pop
        print(meishi_item)
        #return(meishi_item)
    except Exception:
        pass  
    json.dump(meishi_item,f1,ensure_ascii=False,indent=4)
    #result_save(meishi_item)
    f1.close()
```

```python
def detail_page():
    Folder_Path = r'/Users/heweiyuan/py/COM5508_Analytics/Group_project/new_trial_recipe/recipe_url_3' 
    #修改当前工作目录
    os.chdir(Folder_Path)
    #将该文件夹下的所有文件名存入一个列表
    file_list = os.listdir()
    print(file_list)
    file_list.remove('.DS_Store')
    for i in range(0,len(file_list)):
        df = open( Folder_Path + '/'+ file_list[i],encoding='utf-8')
        data = json.loads(df.read(), encoding='utf-8')
        print(len(data))
        for i in range(len(data)):
            filename = data[i]['title']
            fileurl = data[i]['url']
            filetype = data[i]['type']
            filepop = data[i]['popular']
            #detail_page(fileurl, filetype)
            meishi_item = {}
            url = fileurl
            response = request.urlopen(url)
            #response_1 = requests.get(url).content
            #result_save(meishi_item)
            print(fileurl)
            main_ingredient=[]
            sup_ingredient=[]
            sup_vol =[]
            #url = f_url
            try:
                result = response.read().decode('utf-8')
                tree = etree.HTML(result)
                meishi_item['name'] = tree.xpath('//*[@id="tongji_title"]//text()')
                meishi_item['gy'] = tree.xpath('//*[@id="tongji_gy"]//text()')
                meishi_item['people'] = tree.xpath('//*[@id="tongji_rsh"]//text()')
                meishi_item['taste'] = tree.xpath('//*[@id="tongji_kw"]//text()')
                meishi_item['cooking_time'] = tree.xpath('//*[@id="tongji_prsj"]//text()')
                meishi_item['type']= [filetype]
                meishi_item['popular']= filepop
                html = urlopen(url).read().decode("utf-8")
                soup = BeautifulSoup(html,'lxml')
                zhuliao_box = soup.find_all('div',attrs={'class':'yl zl clearfix'})
                for zl in zhuliao_box:
                    for a in zl.find_all('a'):
                        #print(a.text.strip(' '))
                        main_ingredient.append(a.text.strip(' '))
                fuliao_box = soup.find_all('div',attrs={'class':'yl fuliao clearfix'})
                for fl in fuliao_box:
                    for b in fl.find_all('h4'):
                        #print(a.text.strip(' '))
                        sup_ingredient.append(b.text.strip(' '))
                    for b_1 in fl.find_all('span'):
                        #print(b_1.text.strip(''))
                        sup_vol.append(b_1.text.strip(''))
                #a= soup.find_all("div",attrs={'class':'measure'})
                print(main_ingredient)
                #meishi_item['title']= f_name
                meishi_item['ingredient']= str(main_ingredient+sup_ingredient)
                #meishi_item['sup_vol']= sup_vol
                path = '/Users/heweiyuan/py/COM5508_Analytics/Group_project/new_trial_recipe/recipe_2.csv'
                print(meishi_item)
                with open(path,'a+') as csvfile:
                    fieldnames = ['name','gy','people','taste','cooking_time','type','popular','ingredient']
                    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
                    writer.writerow(meishi_item)
                #return(meishi_item)
                #result_save(meishi_item)
            except Exception:
                pass  
```
